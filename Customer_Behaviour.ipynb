```python
# âœ… Import the necessary libraries
import pandas as pd                # For reading and handling data
import seaborn as sns              # For beautiful graphs
import matplotlib.pyplot as plt    # For plotting charts

# âœ… Load the dataset (file must be in same folder)
df = pd.read_csv("Mall_Customers.csv")
print("âœ… Dataset Loaded!")

# âœ… Rename columns to make it easier to use
df.columns = ['CustomerID', 'Gender', 'Age', 'AnnualIncome', 'SpendingScore']

# âœ… Show the first 5 rows of the data
print("ðŸ“Š First 5 rows:")
print(df.head())
```

Data summary and shape

```python
# âœ… Check basic information (types, nulls)
print("ðŸ“¦ Dataset Info:")
print(df.info())

# âœ… Summary statistics (mean, min, max, etc.)
print("\nðŸ“ˆ Summary Stats:")
print(df.describe())
```

Gender Bar Graph

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

Age Distribution(Histogram)

```python
sns.histplot(df['Age'], kde=True, bins=10)
plt.title("Age Distribution")
plt.show()
```

Annual Income distribution

```python
sns.histplot(df['AnnualIncome'], kde=True, color='green')
plt.title("Annual Income Distribution")
plt.show()
```

Spending Score Distribution

```python
sns.histplot(df['SpendingScore'], kde=True, color='orange')
plt.title("Spending Score Distribution")
plt.show()
```

Scatter Plot: Income vs Spending (Colored by Gender)

```python
sns.scatterplot(x='AnnualIncome', y='SpendingScore', hue='Gender', data=df)
plt.title("Income vs Spending Score by Gender")
plt.show()
```

Pairplot â€“ All Numeric Columns Together

```python
sns.pairplot(df[['Age', 'AnnualIncome', 'SpendingScore']])
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

Violin Plot â€“ Income by Gender

```python
sns.violinplot(x='Gender', y='AnnualIncome', hue='Gender', data=df, palette="Set2", legend=False)
plt.title("Income by Gender")
plt.show()
```

Voilin Plot -spending score

```python
sns.violinplot(x='Gender', y='SpendingScore', hue='Gender', data=df, palette="Set1", legend=False)
plt.title("Spending Score by Gender")
plt.show()
```
